```python
import pandas as pd
```

```python
#load in dataset for precipitaiton (global)
df = pd.read_csv('average-precipitation-per-year.csv')
```

```python
#first 5 lines of data
df.head()
```

```python
#summary of data focus on annual precipiatation
df.describe()
```

```python
# check for null values
df.isnull().sum()
```

```python
# remove country code as it is not needed
df = df[['Entity','Year','Annual precipitation']]
```

```python
# first 5 rows of updated dataframe
df.head()
```

```python
# check new df for null values
df.isnull().sum()
```

```python
# check data types for new df
df.dtypes
```

```python
# rename columns to match other datasets
df.rename(columns={
    'Entity':'Country',
    'Annual precipitation':'Annual_precipitation'
}, inplace=True)
```

```python
# checking new column names
df.head()
```

```python
# filtering for years between 2000-2024
df = df[df['Year'].between(2000, 2024)]
```

```python
# output to .csv file
df.to_csv("Yearly_Dengue_Data.csv", index=False)
```
